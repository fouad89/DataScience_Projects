```python
# violin visualization
def violin_vis(x,y,df=customer_df):
    # function to take the x,y axis and produce a violin plot
    plt.style.use('seaborn-pastel')
    plt.figure(figsize=(16,12))
    sns.despine(offset=10, trim=True);
    sns.catplot(x=x,y=y, data=df, kind='violin')
    plt.title(f'{y} Vs. {x}'.title());
    plt.tight_layout()
```

```python
violin_vis(x='Churn',y='tenure')
```

```python
violin_vis(x='Churn',y='MonthlyCharges')
```

```python
def corr_matrix(df):
    corr_mat = df.corr()
    mask = np.zeros_like(corr_mat)
    mask[np.triu_indices_from(mask)] = True
    sns.heatmap(corr_mat,annot=True,mask=mask)
    plt.title("Correlation Matrix")
    plt.xticks(rotation=45)
    plt.yticks(rotation=45)
    plt.tight_layout()
```

```python
corr_matrix(customer_df)
```

## Preprocessing

```python
customer_df.head()
```

```python
customer_df.dtypes
```

```python
# removing empty rows from the Total charges column
customer_df = customer_df[customer_df.TotalCharges != " "]

#check changes
customer_df[customer_df.TotalCharges == " "]
```

```python
customer_df.loc[:,'TotalCharges'] = pd.to_numeric(customer_df.TotalCharges)
```

```python
customer_df.TotalCharges.dtypes
```

```python
customer_df.drop(["customerID"],axis=1,inplace=True)
```

```python
# isolating columns 
# binary columns
bin_cols = pd.DataFrame([customer_df[x] for x in customer_df.columns if customer_df[x].nunique()==2]).T
```

```python
bin_col_names = bin_cols.columns.to_list()
bin_col_names
```

```python
customer_df.head()
```

```python
qualitative = [col for col in customer_df.columns if (customer_df.dtypes[col]=='object')]
quantitative = [col for col in customer_df.columns if customer_df.dtypes[col]!='object' and (col not in bin_col_names)]
```

```python
qualitative
```

```python
quantitative
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
enocder = LabelEncoder()
```

```python

for col in bin_col_names:
    customer_df[col] = enocder.fit_transform(customer_df[col])

        
        
```

```python
customer_df.head()
```

```python
customer_df.Churn.value_counts()
```

```python
multi_cat_cols = [col for col in qualitative if col not in bin_col_names]
multi_cat_cols
```

```python
# convert other category columns into numerical 
customer_df = pd.get_dummies(data=customer_df, columns=multi_cat_cols)
```

```python
customer_df.head()
```

```python
customer_df.columns
```

```python
# normalizing data
scaler = StandardScaler()
```

```python
scaled_df = pd.DataFrame(scaler.fit_transform(customer_df[quantitative]),
                        columns=quantitative)
```

```python
scaled_df
```

```python
# dropping the original quantitative columns and replacing with normalized ones
customer_df.drop(quantitative, axis=1, inplace=True)
```

```python
customer_df.columns # check
```

```python
customer_df = customer_df.merge(scaled_df,left_index=True, right_index=True,how='left')
```

```python
customer_df.shape
```

```python
# check for null values
customer_df[customer_df.isna().any(axis=1)]
```

```python
# drop na values
customer_df = customer_df.dropna()
```

```python
customer_df[customer_df.isna().any(axis=1)]
```

```python
processed_df = customer_df.copy()
```
